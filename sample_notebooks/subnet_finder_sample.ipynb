# How to use HD Subnetwork Finder

```python
from hdsubnetfinder.kernel.kernel_generator import KernelGenerator
from hdsubnetfinder.subnetwork.sub_network_finder import SubNetworkFinder
from hdsubnetfinder.kernel import kernel_util

import hdsubnetfinder.subnetwork.network_util as util
import json
```

```python
# network_file = 'https://raw.githubusercontent.com/idekerlab/hdsubnetfinder/master/tests/resource/small_network.sif'
network_file = 'http://chianti.ucsd.edu/~kono/ci/data/HumanNet.v1.symbol.sif'

generator = KernelGenerator()

# Create Kernel Object from small network (in SIF.  This may take few seconds.)
kernel = generator.create_kernel(network_file)

# If necessary you can save it in a file
kernel_util.write_kernel(kernel, 'human_net.kernel.symbol.txt')
```

```python
# Check the contents of the Kernel file
import pandas as pd

df = pd.read_csv('kernel_sample.txt', sep='\t')
df.head()
```

```python
sif = util.read_sif(network_file)
finder = SubNetworkFinder(kernel=kernel, network=sif)

query = ["NRAS", "KRAS", "MAPK1"]
subnet1 = finder.get_sub_network(query) # Result is stored in CX format.
```
